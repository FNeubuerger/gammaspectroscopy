```python
import pandas as pd
import os
import numpy as np
# Print the current working directory
print(os.getcwd())
# Read the contents of the file into a pandas DataFrame
file_path = 'databasegamma.txt'
columns = ["Nuclide", "Energy (keV)", "Intensity (%)", "Decay mode",  "Half-life", "Half-life unit", "Effective Intens.(dps)"]

# Use error_bad_lines=False to skip lines with too many fields
df = pd.read_table(file_path, skiprows=25, names=columns, sep='\s+', on_bad_lines='skip', index_col=False)
df['Energy (keV)'] = df['Energy (keV)'].replace('X', np.nan).astype(float)
df['Intensity (%)'] = df['Intensity (%)'].replace('--', np.nan).astype(float)
df['Half-life'] = df['Half-life'].replace('--', np.nan).astype(float)
df
```

```python
df.to_parquet('databasegamma.parquet', index=False)
```
